# Pie Chart

```python
#lets make a hypothetical data of a company
areas=['sales','devlopment','marketing','tech','research']
budget=[1.5,2.5,3.5,6.5,10]
```

```python
#pie is the function of pyplot lib #use autopct if you want to find percentage 0.1 mean 1 digit after decimal point
plt.pie(budget,labels=areas,autopct='%0.1f%%')
```

```python
plt.pie(budget,labels=areas,autopct='%0.1f%%') #if you dont want to see unnecessary data use 
plt.show()
```

```python
x=[11,23,45,23,65,34,24]
y=[10,2,45,6,7,3,7]
```

```python
plt.scatter(x,y) #important function
#here we can also tell what we are shwing in x and y axis using labels
plt.xlabel('age')
plt.ylabel('salary')
#if you want to give title to this graph use 
plt.title('Age VS Salary')
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[233,455,764,245,875,865,345,450]
chor_bazzar=[120,200,300,200,100,150,140,240]
```

```python
plt.plot(product_id,online_price,color='green') #this is a 2d plot using list or numpy array #color to change the color of graph
plt.xlabel('product') 
plt.ylabel('price')
plt.title('product price comparision')

#if you dont want to see unecessary data or techinical terms you can simply use plt.show()
plt.show()
```

```python
#if you want to see the point where the graph meet or where our points meet #use marker parameter
plt.plot(product_id,online_price,color='green',marker='*',markersize=10,linestyle='dashed',linewidth=2) #use D means diamond #this is a 2d plot using list or numpy array #color to change the color of graph
#you can also use linestyle like,'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'


plt.xlabel('product') 
plt.ylabel('price')
plt.title('product price comparision')

#if you dont want to see unecessary data or techinical terms you can simply use plt.show()
plt.show()
```

```python
#if you want to comparision between the graphs #and you want to more the one graph in a single graph
#if you want to see the point where the graph meet or where our points meet #use marker parameter
plt.plot(product_id,online_price,color='green',marker='*',markersize=10,label='online prices')#use D means diamond #this is a 2d plot using list or numpy array #color to change the color of graph
plt.plot(product_id,offline_price,color='red',markersize=10,label='offline prices')#marker defualt
plt.plot(product_id,chor_bazzar,color='blue',marker='+',markersize=10,label='chori ke prices')
plt.xlabel('product') 
plt.ylabel('price')
plt.title('product price comparision')
plt.legend() #you can give the location like upper left or right
#if you dont want to see unecessary data or techinical terms you can simply use plt.show()
plt.show()
#now you can see the different grphs in a single graph
#but if do it directly we will confuse people bcs we dont signifing that which color show what prices so we have use label parameter inside the plot function and legend to make it
```
